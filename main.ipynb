```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('./healthcare-dataset-stroke-data.csv')
```

```python
print(f'Number of rows are {len(data)} and columns are {len(data.columns)}.')
```

```python
keys = data['stroke'].value_counts().keys()
values = data['stroke'].value_counts().values
```

```python
sns.barplot(keys, values)
plt.show()
```

```python
# Code shows gender distribution of people having strokes
onlyStrokes = data[data['stroke']==1]
keys = onlyStrokes['gender'].value_counts().keys()
values = onlyStrokes['gender'].value_counts().values
plt.title('Gender distribution amongst people have stroke!')
plt.xlabel('Gender type')
plt.ylabel('Count')
sns.barplot(keys,values)
plt.show()
```

```python
# Code shows gender distribution of people not having strokes
onlyStrokes = data[data['stroke']==0]
keys = onlyStrokes['gender'].value_counts().keys()
values = onlyStrokes['gender'].value_counts().values
plt.title('Gender distribution amongst people have stroke!')
plt.xlabel('Gender type')
plt.ylabel('Count')
sns.barplot(keys,values)
plt.show()
```

```python
# Code shows gender distribution of all people
# onlyStrokes = data[data['stroke']==0]
keys = data['gender'].value_counts().keys()
values = data['gender'].value_counts().values
plt.title('Gender distribution amongst people have stroke!')
plt.xlabel('Gender type')
plt.ylabel('Count')
sns.barplot(keys,values)
plt.show()
```

```python
onlyStrokes = data[data['stroke']==1]
```

```python
smoke = onlyStrokes['smoking_status'].value_counts()['smokes'] + onlyStrokes['smoking_status'].value_counts()['formerly smoked']
nonSmokers = onlyStrokes['smoking_status'].value_counts()['never smoked']
```

```python
print(f'Precentage of smoking people: {(smoke/len(onlyStrokes)) *100}% and non-smoker are: {(nonSmokers/len(onlyStrokes))*100}%')
```
